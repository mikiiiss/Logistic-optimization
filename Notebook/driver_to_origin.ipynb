```python
from haversine import haversine, Unit
def calculate_displacement(row):
    lat, lng = row['lat'], row['lng']
    if isinstance(row['trip_origin'], str):
        origin_lat, origin_lng = row['trip_origin'].split(',')
        origin_lat, origin_lng = float(origin_lat), float(origin_lng)
    else:
        origin_lat, origin_lng = row['trip_origin'], row['trip_origin']  # or some other default value
    distance = haversine((lat, lng), (origin_lat, origin_lng), unit=Unit.KILOMETERS)
    return distance

# Apply the function to each row and add the result as a new column
df['displacement'] = df.apply(calculate_displacement, axis=1)
df.head(5)

```

```python
import numpy as np
struct_data = data.copy()
non_numeric_columns = list(struct_data.select_dtypes(exclude=[np.number]).columns)

print(non_numeric_columns)
```

```python
import numpy as np

# Impute NaN with column means
for col in struct_data.columns:
    if struct_data[col].isnull().any():
        mean_value = struct_data[col].mean()
        struct_data[col].fillna(mean_value, inplace=True)
```

```python

from causalnex.structure.notears import from_pandas
sm = from_pandas(struct_data)
```

```python
from causalnex.plots import plot_structure, NODE_STYLE, EDGE_STYLE
viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)

viz.toggle_physics(False)
viz.show("03_fully_connected.html")
```

```python
from causalnex.plots import plot_structure, NODE_STYLE, EDGE_STYLE
sm.remove_edges_below_threshold(0.8)
viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)
viz.show("support01_thresholded.html")
```
